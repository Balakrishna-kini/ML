```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix
```

```python
#STEP 1 mANUALLY LOAD THE OLIVETTI FACE DATASET
X=np.load("olivetti_faces.npy")
y=np.load("olivetti_faces_target.npy")
```

```python
#step 2 falatten the images (reshape from 3d to 2d)
X=X.reshape(X.shape[0],-1)
```

```python
#step 3 : Split the dataset into training and testing sets
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2)
```

```python
#step 4: Train the Naive Bayesian Classifier 
gnb = GaussianNB()
gnb.fit(X_train,y_train)
```

```python
#step5 : Prdict on the test set 
y_pred=gnb.predict(X_test)
```

```python
#step6: Evaluate the model
accuracy =accuracy_score(y_test,y_pred)
print(f"Accuracy:{accuracy*100:.2f}%")
```

```python
#step 7:Visualization sometest images with prediction

n_images=5
plt.figure(figsize=(10,4))
for i in range(n_images):
    plt.subplot(1,n_images,i+1)
    plt.imshow(X_test[i].reshape(64,64),cmap='gray')
    plt.title(f"True:{y_test[i]}\nPred: {y_pred[i]}")
    plt.axis('off')
plt.show()    
```
